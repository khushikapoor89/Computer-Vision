```python
import cv2
import numpy as np
```

```python
# Create a black image
img = np.zeros((512, 512, 3))
```

```python
from google.colab.patches import cv2_imshow
```

```python
# Display the blank image
cv2_imshow(img)
```

```python
# Re-initialize the blank image (optional, redundant)
img = np.zeros((512, 512, 3))
cv2_imshow(img)
```

```python
# Draw a rectangle with only the border
cv2.rectangle(img, pt1=(100, 100), pt2=(300, 300), color=(255, 0, 0), thickness=3)
cv2_imshow(img)
```

```python
# Draw a filled rectangle
cv2.rectangle(img, pt1=(100, 100), pt2=(300, 300), color=(255, 0, 0), thickness=-1)
cv2_imshow(img)
```
